```python
import pandas as pd
from pathlib import Path
import numpy as np

# File to Load
homicide_csv = Path("data/homicide_data.csv")

# Read homicide data file and store into Pandas DataFrame
df = pd.read_csv(homicide_csv, encoding='latin1')
df
```

```python
df['state'].unique()
```

```python
df.info()
```

```python
#Wisconsin is WI no wI
df['state'] = df['state'].replace('wI', 'WI')
```

```python
#Some states/cities did not disclose victim names
df =  df.drop(columns = ["victim_last", "victim_first"])
```

```python
# Replace "unknown" values with NaN
df.replace('Unknown', pd.NA, inplace=True)

# Drop rows with NaN values in any column
df.dropna(axis=0, how='any', inplace=True)

# Now, 'homicide_data' contains rows where any column is not "Unknown"
```

```python
df.info()
```

```python
47478/52179  # lost about 10% of the data 
```

```python
df.head()
```

```python
df.disposition = df.disposition.replace("Closed without arrest", "No Arrest")
df.disposition = df.disposition.replace("Open/No arrest", "No Arrest")
df.disposition = df.disposition.replace("Closed by arrest", "Arrest Made")
```

```python
df.disposition.value_counts()
```

```python
df.columns
```

```python
df['reported_date'] = pd.to_datetime(df['reported_date'], format='%Y%m%d')
```

```python
df['reported_year'] = df['reported_date'].dt.year
df['reported_month'] = df['reported_date'].dt.month
df['reported_weekday'] = df['reported_date'].dt.dayofweek
```

```python
df['reported_year'].value_counts()
```

```python
df['reported_month'].value_counts()
```

```python
df['reported_weekday'].value_counts()
```

```python
df.head()
```

```python
# Convert 'victim_age' to numeric
df['victim_age'] = pd.to_numeric(df['victim_age'])

# Define the age bins and labels
age_bins = [0, 18, 30, 45, 64, float('inf')]
age_labels = ['0-17', '18-29', '30-44', '45-64', '65+']

# Bin the 'victim_age' column and create a new column 'age_range'
df['age_range'] = pd.cut(df['victim_age'], bins=age_bins, labels=age_labels, right=False)
df['age_range']
```

```python
df.age_range.value_counts()
```

```python
# Define a function to map month values to seasons
def get_season(reported_month):
    if reported_month in [12,1,2]:
        return 'Winter'
    elif reported_month in [3, 4, 5]:
        return 'Spring'
    elif reported_month in [6, 7, 8]:
        return 'Summer'
    elif reported_month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Unknown'  # Handle invalid month values if any

# Apply the function to create a new "season" column
df['season'] = df['reported_month'].apply(get_season)
df['season']
```

```python
df.season.value_counts()
```

```python
df.head()
```

```python
# Create a new column 'LOCATION' by concatenating 'city' and 'state' with a comma separator
df['LOCATION'] = df['city'] + ', ' + df['state']
df['LOCATION']
```

```python
df['LOCATION'].unique()
```

```python
#There is no Tulsa, AL. The lon and lat indicate Oklahoma.
# Replace 'AL' with 'OK' for 'LOCATION' containing 'city' Tulsa.
df['LOCATION'] = df['LOCATION'].str.replace(r'Tulsa.*, AL$', 'Tulsa, OK', regex=True)
```

```python
locations_to_keep = df['LOCATION'].unique()
locations_to_keep
```

```python
# Specify the path for the new CSV file where you want to save the filtered data
output_csv = Path("data/machine_learning_homicide.csv")

# Save the filtered DataFrame to the new CSV file
df.to_csv(output_csv, index=False)
```

```python
id_vars = ['NAME', 'STNAME', 'LOCATION']
value_vars = ['2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017']
```

```python
df_melt = pd.melt(filtered_pop_df,id_vars=id_vars,value_vars=value_vars,var_name="YEAR", value_name="POPULATION")
df_melt['YEAR'] = df_melt.YEAR.astype(int)
df_melt['POPULATION'] = df_melt.POPULATION.astype(int)
```

```python
df_melt.info()
```

```python
df_final = pd.merge(df, df_melt, left_on = ['LOCATION','reported_year'],
                    right_on = ['LOCATION', 'YEAR'], how = 'left')
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
df_per = df_final.groupby(["LOCATION","reported_year"]).agg({"uid":"size","POPULATION": 'max'}).reset_index()
df_per["homicides_per_100000"] = df_per["uid"]/df_per["POPULATION"]*100000

df_per
```

```python
# Specify the path for the new CSV file where you want to save the filtered data
output_csv = Path("data/filtered_per_100000.csv")

# Save the filtered DataFrame to the new CSV file
df_per.to_csv(output_csv, index=False, header=True)
```
